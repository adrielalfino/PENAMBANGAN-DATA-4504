```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv(r"C:\dota2_hero_stats.csv")
```

```python
print(data.head())
print(data.info())
```

```python
features = data[['Hero', 'Win Rate', 'Pick Rate', 'KDA Ratio', 'Patch']]
```

```python
plt.figure(figsize=(12, 8))
plt.barh(features['Hero'], features['Win Rate'], color='skyblue')
plt.xlabel('Win Rate (%)')
plt.title('Win Rate Hero Dota 2')
plt.gca().invert_yaxis()  # Membalikkan sumbu Y agar hero dengan win rate tertinggi di atas
plt.show()
```

```python
top_10_heroes = features.sort_values(by='Win Rate', ascending=False).head(10)

# Membuat diagram batang
plt.figure(figsize=(10, 6))
plt.barh(top_10_heroes['Hero'], top_10_heroes['Win Rate'], color='skyblue')
plt.xlabel('Win Rate (%)')
plt.title('10 Hero Dota 2 dengan Win Rate Tertinggi')
plt.gca().invert_yaxis()  # Membalikkan sumbu Y agar hero dengan win rate tertinggi di atas
plt.show()
```

```python
top_10_heroes_pick_rate = features.sort_values(by='Pick Rate', ascending=False).head(10)

# Membuat diagram batang
plt.figure(figsize=(10, 6))
plt.barh(top_10_heroes_pick_rate['Hero'], top_10_heroes_pick_rate['Pick Rate'], color='lightgreen')
plt.xlabel('Pick Rate (%)')
plt.title('10 Hero Dota 2 dengan Pick Rate Tertinggi')
plt.gca().invert_yaxis()  # Membalikkan sumbu Y agar hero dengan pick rate tertinggi di atas
plt.show()
```

```python
top_10_heroes_kda = features.sort_values(by='KDA Ratio', ascending=False).head(10)

# Membuat diagram batang
plt.figure(figsize=(10, 6))
plt.barh(top_10_heroes_kda['Hero'], top_10_heroes_kda['KDA Ratio'], color='blue')
plt.xlabel('KDA Ratio')
plt.title('10 Hero Dota 2 dengan KDA Ratio Tertinggi')
plt.gca().invert_yaxis()  # Membalikkan sumbu Y agar hero dengan KDA tertinggi di atas
plt.show()
```
